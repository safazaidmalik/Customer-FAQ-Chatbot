```python
# my functions

def find_similarity(inputs,sentences):
    #find input's sentence embeddings
    input_embeddings = sentence_model.encode(inputs, convert_to_tensor = True)
    #find corpus sentence embeddings
    corpus_embeddings = sentence_model.encode(sentences, convert_to_tensor = True)
    # find cosine-similarities 
    cosine_scores = util.cos_sim(corpus_embeddings, input_embeddings)
    #Find the pair with the highest cosine similarity score
    pair = []
    max_score = -1
    max_index = -1
    for i in range(len(cosine_scores)):
        if cosine_scores[i]>max_score:
            max_score = cosine_scores[i]
            max_index = i
            print('new max score = ', max_score)
    pair = [inputs[0], sentences[max_index]]
    print('most matching sentence: \n', pair)
    print('Similarity:', max_score)
```
