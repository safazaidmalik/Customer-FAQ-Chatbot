```python
import random
import nltk
from nltk.stem import WordNetLemmatizer # to reduce to stem word
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD
```

```python
import random # Choosing the random response
import json
import pickle # for serialization
import numpy as np
```

```python
lemmatizing_points = WordNetLemmatizer()
with (open('data.json')) as fffss:
    intents = json.load(fffss)

AllParts = []
Category = []
findings = []

stopWords = ['?', '!', '.', ',',"!","#","$","%","&","'","(",")","~"]
```

```python

for intent in intents['intents']:
    for random in intent['patterns']:
        WordInList = nltk.word_tokenize(random)
        findings.extend(WordInList)
        AllParts.append((WordInList, intent['tag']))
        if intent['tag'] not in Category:
            Category.append(intent['tag'])
            
print(findings)

```

```python
new_findings=[]
for point in findings:
    if point not in stopWords:
        new_findings.append(lemmatizing_points.lemmatize(point))
        
        
findings = sorted(set(new_findings))

Category = sorted(set(Category))
training = []
final = [0] * len(Category)

pickle.dump(findings, open('queries.pkl','wb'))
pickle.dump(Category, open('tags.pkl', 'wb'))

print(findings)
```

```python
for part in AllParts:
    container = []
    point_order = part[0]
    point_order = [lemmatizing_points.lemmatize(point.lower()) for point in point_order]
    for point in findings:
        container.append(1) if point in point_order else container.append(0)
    
    check = list(final)
    check[Category.index(part[1])] = 1
    training.append([container, check])
```

```python

training = np.array(training)


train_x = list(training[:, 0])
train_y = list(training[:, 1])


model = Sequential()
model.add(Dense(128, input_shape=(len(train_x[0]), ), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(len(train_y[0]), activation='softmax'))

sgd = SGD(learning_rate=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
ai_model = model.fit(np.array(train_x), np.array(train_y), epochs=250, batch_size=3, verbose=1)
model.save('Model.h5', ai_model)
print("AI Chatbot Model Built")
```
